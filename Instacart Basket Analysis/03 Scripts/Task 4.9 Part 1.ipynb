```python
#Importing libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#Importing data
path = r'C:\Users\dylan\Documents\Instacart Basket Analysis'

customers = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'customers.csv'))
customers.head()
```

```python
#Check for mixed type data
for col in customers.columns.tolist():
    weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis=1)
    if len (customers[weird]) > 0:
        print(col)
```

```python
#Convert all to one value
customers['First Name'] = customers['First Name'].astype('str')
```

```python
#Checking for missing values
customers.isnull().sum()
```

```python
#Checking for duplicated values
customers_dups = customers[customers.duplicated()]
customers_dups
```

```python
#Importing other data for combining
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_flags.pkl'))
```

```python
ords_prods_merge.head()
```

```python
customers_merged = ords_prods_merge.merge(customers, on = 'user_id', indicator = 'exist')
```
