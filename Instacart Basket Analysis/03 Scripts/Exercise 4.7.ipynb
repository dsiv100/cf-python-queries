```python
#Importing libraries
import pandas as pd
import numpy as np
import os
```

```python
#Importing data
path = r'C:\Users\dylan\Documents\Instacart Basket Analysis\02 Data'

df_ords_prods = pd.read_pickle(os.path.join(path, 'Prepared Data', 'ords_prods_merge.pkl'))
df_ords_prods.head()
```

```python
df = df_ords_prods[:1000000]
```

```python
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
df['prices'].max()
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#New for loop to check for two days
new_result = []

for value in df_ords_prods['order_dow']:
    if value == 0 or value == 1:
        new_result.append('Busiest days')
    elif value == 4 or value == 3:
        new_result.append('Least busy')
    else:
        new_result.append('Regularly busy')

new_result
```

```python
#Checking frequency of order hour
df_ords_prods['hour_of_order'].value_counts(dropna=False)
```

```python
#For loop to filter hours
hour_result = []

for value in df_ords_prods['hour_of_order']:
    if value == 10:
        hour_result.append('Most orders')
    elif value == 3:
        hour_result.append('Fewest orders')
    else: hour_result.append('Average orders')
```

```python
#Printing frequency
df_ords_prods['busiest_period_of_day'] = hour_result
df_ords_prods['busiest_period_of_day'].value_counts(dropna=False)
```
