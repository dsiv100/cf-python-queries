```python
#Importing libraries
import pandas as pd
import numpy as np
import os
```

```python
#Importing data
path = r'C:\Users\dylan\Documents\Instacart Basket Analysis\02 Data'

ords_prods_merge = pd.read_pickle(os.path.join(path, 'Prepared Data', 'ords_prods_merge_4_8.pkl'))
ords_prods_merge.head()
```

## Loyalty flag done in earlier task


```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

## It seem s like loyal customers actually have less average spending, likely because they're ordering so frequently.

```python
#Creating column to track spending
ords_prods_merge['avg_prod_price'] = ords_prods_merge.groupby('user_id').agg({'prices' : ['mean']})
```

```python
ords_prods_merge.head()
```

```python
#Adding flag for spending
ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
#Creating order frequency flag
ords_prods_merge['median_days'] = ords_prods_merge.groupby('user_id').agg({'days_since_prior_order' : ['median']})
ords_prods_merge.head()
```

```python
#adding frequency flag
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'

ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

```python
#Exporting data
ords_prods_merge.to_pickle(os.path.join(path, 'Prepared Data', 'ords_prods_flags.pkl'))
```
